```python
# obama.jpg에 포함된 얼굴 영역의 임베딩 벡터 구하기
# get_gropped_face() 활용
# face recognition 패키지에서 제공하는 얼굴 임베딩 벡터 구하는 함수는
## face_recognition.face_encodings()
image_file = os.path.join(dir_path, '태연.jpg')
face = get_gropped_face(image_file) # 얼굴영역 구하는 함수

# 얼굴 영역 갖고 얼굴 임베딩 벡터 구하는 함수
def get_face_embedding(face):
    return face_recognition.face_encodings(face)

embedding = get_face_embedding(face)
embedding
```

```python
# images 디렉토리 안에 있는 모든 이미지 파일의 임베딩을 구해서 dict 구조에 담아 리턴하는 함수 만들기
# 사람이름: 임베딩벡터
# 함수이름은 get_face_embedding_dict(dir_path)

def get_face_embedding_dict(dir_path):
    file_list = os.listdir(dir_path)
    embedding_dict = {}
    
    for file in file_list:
        img_path = os.path.join(dir_path, file)
        face = get_gropped_face(img_path)
#         if len(face) == 0: continue
        embedding = get_face_embedding(face)
        if len(embedding) > 0: # 얼굴영역이 제대로 detect되지 않으면 len(embedding) == 0인 경우가 종종 발생
            embedding_dict[os.path.splitext(file)[0]] = embedding[0]
            # os path.splitext(file)[0]에는 이미지파일명에서 확장자를 제거한 이름이 담김
            
    return embedding_dict
```
